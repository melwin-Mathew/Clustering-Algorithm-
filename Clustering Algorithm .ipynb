```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import altair as alt

# Load the Iris dataset
iris = load_iris()

# Convert to a Pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Drop the species column
X = df.drop('species', axis=1, errors='ignore')

# Determine the optimal number of clusters using the elbow method
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the elbow method graph
elbow_data = pd.DataFrame({'k': range(1, 11), 'inertia': inertia})
chart = alt.Chart(elbow_data).mark_line(point=True).encode(
    x='k:O',
    y='inertia:Q',
    tooltip=['k', 'inertia']
).properties(
    title='Elbow Method for Optimal k'
)
chart.save('kmeans_elbow_method.json')

# Apply KMeans clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_clusters = kmeans.fit_predict(X)

# Add cluster labels
df['kmeans_cluster'] = kmeans_clusters

# Visualize the clusters
chart = alt.Chart(df).mark_circle().encode(
    x='sepal length (cm)',
    y='sepal width (cm)',
    color='kmeans_cluster:N',
    tooltip=iris.feature_names + ['kmeans_cluster']
).properties(
    title='KMeans Clustering'
)
chart.save('kmeans_clusters.json')

# Apply hierarchical clustering with the single method
single_clustering = AgglomerativeClustering(n_clusters=3, linkage='single')
single_clusters = single_clustering.fit_predict(X)

# Add cluster labels
df['single_cluster'] = single_clusters

# Visualize the clusters
chart = alt.Chart(df).mark_circle().encode(
    x='sepal length (cm)',
    y='sepal width (cm)',
    color='single_cluster:N',
    tooltip=iris.feature_names + ['single_cluster']
).properties(
    title='Hierarchical Clustering (Single)'
)
chart.save('hierarchical_clusters_single.json')

# Apply hierarchical clustering with the complete method
complete_clustering = AgglomerativeClustering(n_clusters=3, linkage='complete')
complete_clusters = complete_clustering.fit_predict(X)

# Add cluster labels
df['complete_cluster'] = complete_clusters

# Visualize the clusters
chart = alt.Chart(df).mark_circle().encode(
    x='sepal length (cm)',
    y='sepal width (cm)',
    color='complete_cluster:N',
    tooltip=iris.feature_names + ['complete_cluster']
).properties(
    title='Hierarchical Clustering (Complete)'
)
chart.save('hierarchical_clusters_complete.json')
chart.display()

# Apply hierarchical clustering with the average method
average_clustering = AgglomerativeClustering(n_clusters=3, linkage='average')
average_clusters = average_clustering.fit_predict(X)

# Add cluster labels
df['average_cluster'] = average_clusters

# Visualize the clusters
chart = alt.Chart(df).mark_circle().encode(
    x='sepal length (cm)',
    y='sepal width (cm)',
    color='average_cluster:N',
    tooltip=iris.feature_names + ['average_cluster']
).properties(
    title='Hierarchical Clustering (Average)'
)
chart.save('hierarchical_clusters_average.json')
chart.display()

# Apply hierarchical clustering with the ward method
ward_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')
ward_clusters = ward_clustering.fit_predict(X)

# Add cluster labels
df['ward_cluster'] = ward_clusters

# Visualize the clusters
chart = alt.Chart(df).mark_circle().encode(
    x='sepal length (cm)',
    y='sepal width (cm)',
    color='ward_cluster:N',
    tooltip=iris.feature_names + ['ward_cluster']
).properties(
    title='Hierarchical Clustering (Ward)'
)
chart.save('hierarchical_clusters_ward.json')
chart.display()

warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
#  KMeans clustering might be suitable for the Iris dataset because it assumes that clusters are spherical and of similar size. The Iris dataset, although containing different species, has features (sepal and petal dimensions) that could potentially form such clusters. However, KMeans might not perform well if the clusters are of varying shapes or densities.

```
